# SVM(Support Verctor Machine)

```python
from sklearn import datasets
from sklearn.svm import SVC
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
c = [ ['r','g','b'][i] for i in y] # 각 레이블 값에 따라 색상을 지정하는 리스트
```

```python
X[0]
```

```python
y[0]
```

```python
c[0:10]
```

```python
import pandas as pd

%matplotlib inline
df = pd.DataFrame(X,columns=iris.feature_names) #feature_names를 입력하면 iris의 요소를 볼 수 있다
df.plot(x=1, y=2, c=c, kind='scatter')
```

```python
svc = SVC(kernel='linear') #kernel을 이용하여 선형적으로 설정
svc.fit(X,y) #fit은 학습을 하는 데이터
```

```python
svc.predict(X)
```

```python
svc.score(X,y)
```

```python
#테스트, 스코어, 학습 데이터를 일정한 비율로 골고루 섞어서(random) 결과값을 내야함
```

# RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=1000) # default 10. 복잡한 데이터에서 더 큰 값 사용
rfc.fit(X,y)
```

```python
rfc.predict(X)
```

```python
rfc.score(X,y)
```
