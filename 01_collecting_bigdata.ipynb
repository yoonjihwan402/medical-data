```python
import requests, sys

# ìš”ì²­ url ì„¤ì •
server_url = "https://rest.ensembl.org"
ext_url = "/overlap/id/ENSG00000157764?feature=gene"

# HTTP GET method í˜¸ì¶œ
r = requests.get(server_url + ext_url, headers={ "Content-Type" : "application/json"})

# ì˜ˆì™¸ ì²˜ë¦¬
if not r.ok:
  r.raise_for_status()
  sys.exit()

# Response í™•ì¸
decoded = r.json()
print(repr(decoded))
```

```python
# ë°ì´í„° í”„ë ˆì„ í˜•íƒœë¡œ ì €ì¥ ë° ì¶œë ¥
import pandas as pd
df = pd.DataFrame(decoded)
df
```
